## Fact Table creation

```sql
-- first our master table is ordersdwh.stg_orders_transformed , by using this we will create the fact table
select * from ordersdwh.stg_orders_transformed;
```

```sql
-- first create the fact table
create or replace table ordersdwh.factorders
(
  orderid int, 
  quantity int, 
  unitprice decimal,
  totalamount decimal,
  dimorderkey int,
  dimproductkey int,
  dimregionkey int,
  dimcustomerkey int
);
```

```sql
-- for one dimensional with dimcustomers, first check what is in dimcustomers table
select * from ordersdwh.dimcustomers;
```

```sql
-- fact table with one dimension(customerdim)
select t.OrderID,t.Quantity,t.UnitPrice,t.TotalAmount,c.dimcustomerkey
from ordersdwh.stg_orders_transformed t
left join ordersdwh.dimcustomers c
on t.CustomerID = c.customerid;
```

```sql
-- now fact table with dimcustomers and productdim
select t.OrderID,t.Quantity,t.UnitPrice,t.TotalAmount,c.dimcustomerkey,p.dimproductkey
from ordersdwh.stg_orders_transformed t
left join ordersdwh.dimcustomers c
on t.CustomerID = c.customerid
left join ordersdwh.dimproducts p
on t.ProductID = p.productid;
```

```sql
-- now check what is in dimregions table
select * from ordersdwh.dimregions;
```

```sql
-- now fact table with dimcustomers,productdim and regiondim
select t.OrderID,t.Quantity,t.UnitPrice,t.TotalAmount,c.dimcustomerkey,p.dimproductkey,r.dimregionkey
from ordersdwh.stg_orders_transformed t
left join ordersdwh.dimcustomers c
on t.CustomerID = c.customerid
left join ordersdwh.dimproducts p
on t.ProductID = p.productid
left join ordersdwh.dimregions r
on t.RegionID = r.regionid;
-- now insert the data into factorders table
```

```sql
-- now check, what is in ordersdim table
select * from ordersdwh.dimorders;
```

```sql
-- now fact table with dimcustomers,productdim,regiondim and orderdim
select t.OrderID,t.Quantity,t.UnitPrice,t.TotalAmount,c.dimcustomerkey,p.dimproductkey,r.dimregionkey,o.dimorderkey
from ordersdwh.stg_orders_transformed t
left join ordersdwh.dimcustomers c
on t.CustomerID = c.customerid
left join ordersdwh.dimproducts p
on t.ProductID = p.productid
left join ordersdwh.dimregions r
on t.RegionID = r.regionid
left join ordersdwh.dimorders o
on t.orderdate = o.orderdate;
-- 
```
